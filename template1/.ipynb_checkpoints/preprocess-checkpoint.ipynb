# Import Library

```python
import pandas as pd
import numpy as np
```

# Import Raw Data

```python
df_customer = pd.read_csv("../data/temp1/input/df_dummy_ecom_customer.csv")
df_trans = pd.read_csv("../data/temp1/input/df_dummy_ecom_transaction.csv")
df_product = pd.read_csv("data/input/df_dummy_ecom_product.csv")
```

# Display Raw Data

```python
df_customer.head(2)
```

```python
df_trans.head(2)
```

```python
df_product.head(2)
```

# Fix Missing Data

```python
print('1. ------------------------------------------------')
display(df_customer.isnull().sum())
print('2. ------------------------------------------------')
display(df_trans.isnull().sum())
print('3. ------------------------------------------------')
display(df_product.isnull().sum())
```

```python
print('1. ------------------------------------------------')
display(df_customer.loc[df_customer['customer_id'].isnull()])
print('2-1. ------------------------------------------------')
display(df_trans.loc[df_trans['product_id'].isnull()])
print('2-2. ------------------------------------------------')
display(df_trans.loc[df_trans['cancellation_date'].isnull()])
```

## Remove the rows with missing values

```python
# Remove the missing value
df_customer.dropna(subset=['customer_id'], inplace=True)
df_trans.dropna(subset=['product_id'], inplace=True)

# Check the result
print('1. ------------------------------------------------')
display(df_customer.isnull().sum())
print('2. ------------------------------------------------')
display(df_trans.isnull().sum())
print('3. ------------------------------------------------')
display(df_product.isnull().sum())
```

# Fix Duplicated Data

```python
print(df_customer.duplicated(subset=['customer_id']).sum())
print(df_trans.duplicated(subset=['transaction_id']).sum())
print(df_product.duplicated(subset=['product_id']).sum())
```

```python
df_trans.loc[df_trans.duplicated(subset=['transaction_id'],keep=False)].sort_values(['transaction_id'])
```

## Remove the duplicated data

```python
df_trans.drop_duplicates(subset=['transaction_id'], keep='last', inplace=True)
# Check the result
df_trans.duplicated(subset=['transaction_id']).sum()
```

# Combine the Data Frames

```python
df_merge = pd.merge(df_trans, df_customer, on='customer_id', how='left')
df_merge = pd.merge(df_merge, df_product, on='product_id', how='left')
df_merge.head()
```

# Create Daily Sales Data

```python
df_daily_sales = df_merge.groupby('transaction_date', as_index=False).agg({'sales':'sum'})
df_daily_sales
```

# Create Daily Sales Detail Data

```python
df_daily_sales_detail = df_daily_sales.copy()
df_daily_sales_detail['transaction_date']=pd.to_datetime(df_daily_sales['transaction_date'])
df_daily_sales_detail['year'] = df_daily_sales_detail['transaction_date'].dt.year
df_daily_sales_detail['month'] = df_daily_sales_detail['transaction_date'].dt.month
df_daily_sales_detail['week'] = df_daily_sales_detail['transaction_date'].dt.isocalendar().week
df_daily_sales_detail['day'] = df_daily_sales_detail['transaction_date'].dt.day_name()

df_daily_sales_detail
```

# Output the Data

```python
df_customer.to_csv("data/output/customer_processed.csv", index=False)
df_trans.to_csv("data/output/transaction_processed.csv", index=False)
df_product.to_csv("data/output/product_processed.csv", index=False)
df_merge.to_csv("data/output/merge.csv", index=False)
df_daily_sales.to_csv("data/output/daily_sales.csv", index=False)
df_daily_sales_detail.to_csv("data/output/daily_sales_detail.csv", index=False)
```
